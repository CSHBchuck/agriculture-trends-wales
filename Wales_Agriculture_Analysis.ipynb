```python
import pandas as pd

data_path = "/Users/williambroderick/Documents/Cristina/Wales_Agriculture_Portfolio/Data"
file_name = "survey-agriculture-and-horticulture-june-2025-168778.ods"
full_path = f"{data_path}/{file_name}"

# Skip one more row to get clean data
df_hist = pd.read_excel(
    full_path, 
    sheet_name="Historical_Series", 
    engine="odf",
    skiprows=5,      # Changed from 4 to 5
    header=0
)

print("Shape:", df_hist.shape)
print("\nFirst 8 rows:")
display(df_hist.head(8))
print("\nColumn names (first 10):")
print(df_hist.columns.tolist()[:10])
print("...")
print("Last 5 columns:")
print(df_hist.columns.tolist()[-5:])
```

```python
# Remove rows that are completely empty or belong to the next table
df_land = df_hist.dropna(how='all')                    # Remove fully empty rows
df_land = df_land[~df_land.iloc[:, 0].astype(str).str.contains('Table|Livestock', na=False)]

print("Cleaned shape:", df_land.shape)
print("\nCleaned data:")
display(df_land)
```

```python
# Separate the two tables
df_land_use = df_hist.iloc[0:5].copy()           # First 5 rows = Land use
df_livestock = df_hist.iloc[8:12].copy()         # Rows 8-11 = Livestock

print("=== LAND USE ===")
print(df_land_use[['Land use (in hectares)', '2015', '2020', '2025']])

print("\n=== LIVESTOCK ===")
print(df_livestock[['Land use (in hectares)', '2015', '2020', '2025']])
```

```python
import pandas as pd

# Rename the first column
df_livestock_clean = df_livestock.rename(columns={'Land use (in hectares)': 'Livestock Category'})

# Reshape from wide to long format
df_long = df_livestock_clean.melt(
    id_vars=['Livestock Category'], 
    var_name='Year', 
    value_name='Number'
)

# Convert Year to number
df_long['Year'] = pd.to_numeric(df_long['Year'], errors='coerce')

# Keep data from 1990 onwards
df_recent = df_long[df_long['Year'] >= 1990].copy()

print("Reshaped Livestock data from 1990 onwards:")
display(df_recent.head(20))
print("\nTotal rows:", len(df_recent))
print("\nYears covered:", df_recent['Year'].min(), "to", df_recent['Year'].max())
```

```python
# Calculate percentage change from 1990 to 2025
summary = df_recent.groupby('Livestock Category').agg(
    Number_1990=('Number', lambda x: x[df_recent.loc[x.index, 'Year'] == 1990].values[0] if any(df_recent.loc[x.index, 'Year'] == 1990) else None),
    Number_2025=('Number', lambda x: x[df_recent.loc[x.index, 'Year'] == 2025].values[0] if any(df_recent.loc[x.index, 'Year'] == 2025) else None)
).reset_index()

summary['Change_Absolute'] = summary['Number_2025'] - summary['Number_1990']
summary['Change_Percent'] = ((summary['Number_2025'] - summary['Number_1990']) / summary['Number_1990'] * 100).round(1)

print("Livestock Changes in Wales (1990 → 2025):")
display(summary.sort_values('Change_Percent', ascending=False))
```

```python
# Filter only 2015 and 2025
df_2015_2025 = df_recent[df_recent['Year'].isin([2015, 2025])].copy()

# Pivot to have years as columns
pivot = df_2015_2025.pivot(index='Livestock Category', columns='Year', values='Number').reset_index()
pivot.columns = ['Livestock Category', 'Number_2015', 'Number_2025']

# Calculate changes
pivot['Change_Absolute'] = pivot['Number_2025'] - pivot['Number_2015']
pivot['Change_Percent'] = ((pivot['Number_2025'] - pivot['Number_2015']) / pivot['Number_2015'] * 100).round(1)

# Sort by percentage change
pivot = pivot.sort_values('Change_Percent', ascending=False)

print("Livestock Changes in Wales (2015 → 2025):")
display(pivot)
```
